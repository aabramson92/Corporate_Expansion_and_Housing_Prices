```python
vaa = pd.read_csv('/Users/alexy161/Downloads/AnnouncementsData.csv')
vaa.drop(vaa.index[2688:],inplace=True)
vaa.columns = ['company', 'county', 'NAICS', 'descript', 'date', 'new_expand', 'jobs_added', 'investment']
vaa.jobs_added = vaa.jobs_added.astype(int)
vaa.NAICS = vaa.NAICS.astype(int)
vaa.date = vaa.date.apply(dateutil.parser.parse)
vaa['month'] = pd.to_datetime(vaa['date']).map(lambda dt: dt.replace(day=1))
vaa['agriculture_added'] = np.where(vaa.NAICS < 100000, vaa.jobs_added, 0)
vaa['manufacturing_added'] = np.where((vaa.NAICS >= 100000) & (vaa.NAICS < 400000), vaa.jobs_added, 0)
vaa['service_added'] = np.where(vaa.NAICS >= 400000, vaa.jobs_added, 0)
vaa['new'] = np.where(vaa.new_expand == 'N', vaa.jobs_added, 0)
vaa['expansion'] = np.where(vaa.new_expand == 'E', vaa.jobs_added, 0)
vaa.NAICS = vaa.NAICS.astype(str)
vaa1 = vaa.groupby(by=['month', 'county']).sum().reset_index()
```
